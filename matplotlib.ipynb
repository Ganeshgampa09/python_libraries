#matplotlib
This library is used for visualization.

In matplotlib we have:-

1)pie()

2)scatter()

3)plot()

3)bar()

4)barh()

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
departments=['development','HR','customer support','Devops']
budget=[5,2,1,9]
```

```python
plt.pie(budget,labels=departments)
```

##when we use this plt.show() we can directly see bar graph itself.

```python
plt.pie(budget,labels=departments)
plt.show()
```

```python
plt.pie(budget,labels=departments,autopct="%0.2f%%")
plt.show()
```

```python
x1=[20,33,88,99,33,46,78,67,82]
x2=[22,34,56,67,89,91,36,76,71]

plt.scatter(x1,x2)
plt.xlabel('age')
plt.ylabel('salary')
plt.title('age vs salary')
```

```python
product_id=[1,2,3,4,5,6,7,8]
product_prices=[222,333,432,543,342,657,789,675]
```

#xlabel and ylabel.
like when we use this xlabel and ylabel we can give names to the xaxis and yaxis. we can use this title() to write heading on the bar graph like what is it about.

```python
plt.plot(product_id,product_prices,color='red',marker='D',markersize=10)
plt.xlabel('product_id')
plt.ylabel('product_prices')
plt.title('id vs prices')
```

```python
product_id=[1,2,3,4,5,6,7,8]
product_prices=[222,333,432,543,342,657,789,675]
offline_price=[121,343,546,743,521,547,879,345]
chormarket_price=[101,201,308,221,543,350,65,230]
```

In this plot we have used **marker** this is used for marking the exact value in the plot.so that when somebody see that he can easily know the exact value.
And also we have used this legend() it represents the lines in our plot.It adds a small box automatically based upon the empty space in the plot.

```python
plt.plot(product_id,product_prices,marker='D',markersize=10,label="online_price")
plt.plot(product_id,offline_price,markersize=10,label="offline_price")
plt.plot(product_id,chormarket_price,marker='*',label="chormarket_price")
plt.xlabel('product_id')
plt.ylabel('product_prices')
plt.title('id vs prices')
plt.legend()
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017]
amount=[10000,15000,12000,13000,21000,22000]
plt.bar(year,amount,color='red')
plt.xlabel('year')
plt.ylabel('amount')
plt.show()
plt.title("year vs amount")
```
